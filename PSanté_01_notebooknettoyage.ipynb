```python
FR_food_filter3.describe()
```

#### Interquartile range

```python

Q1 = FR_food_filter3.quantile(q=.25)
Q3 = FR_food_filter3.quantile(q=.75)
IQR=Q3-Q1

outliers= (FR_food_filter3 < (Q1-1.5*IQR)) | (FR_food_filter3 > (Q3+1.5*IQR))
outliers.sum()
```

#### Suppression des valeurs aberrantes

```python
FR_food_no_out = FR_food_filter3[~((FR_food_filter3 < (Q1-1.5*IQR)) | (FR_food_filter3 > (Q3+1.5*IQR))).any(axis=1)]
FR_food_no_out.shape
```

```python
FR_food_no_out.describe()
```

```python
sns.set(font_scale=1.7)
sns.boxplot(data=FR_food_no_out, showmeans=True,orient='h')
from pylab import rcParams
rcParams['figure.figsize'] = 25,30

plt.suptitle("Quantitative variables")
```
